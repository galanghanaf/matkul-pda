## Kode untuk Mengambil Subset Data berdasarkan Rentang Tanggal.

```python
tanggal_awal = pd.to_datetime("2005-11-20")
tanggal_akhir = pd.to_datetime("2021-03-13")
country = ["Asia", "Europe", "America", "Online"]
subset_df = df[(df["StartDate"] >= tanggal_awal) & (df["StartDate"] <= tanggal_akhir)]
```

## Kode untuk Melakukan Reindex pada Kolom Tanggal.

```python
total_prize = subset_df["TotalUSDPrize"].sum()
rata_rata_prize = subset_df["TotalUSDPrize"].mean()
print(f"Total pengeluaran: {total_prize:,}")
print(f"Rata-rata pengeluaran: {rata_rata_prize:,}")
```

```python
subset_df.groupby("StartDate").sum().sort_values("TotalUSDPrize", ascending=False)
```

## Kode untuk Menghitung Total dan Rata-Rata Pengeluaran.

```python
line_df = subset_df.groupby("StartDate").sum()
line_df = line_df.reindex(pd.date_range(start=tanggal_awal, end=tanggal_akhir,
freq="1D"))
line_df = line_df.fillna(0)
line_df = line_df.reset_index()
fig = px.line(line_df, x="index", y="TotalUSDPrize")
fig.show()
```

```python
pie_df = subset_df.groupby("Region").sum().reset_index()
fig = px.pie(pie_df, names="Region", values="TotalUSDPrize")
fig.show()
```

## Kode untuk Menampilkan Tabel Pivot antara Tanggal, Region, dan Total Hadiah.

```python
subset_df.pivot_table(index="StartDate", columns="Region", values="TotalUSDPrize",
fill_value="-", margins=True, margins_name="Grand Total")
```
